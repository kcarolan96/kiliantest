```python
# Directory containing the tfrecord files
tfrecord_dir = "data/tfrecords/train"

# Get list of file paths
tfrecord_filenames_train = [os.path.join(tfrecord_dir, f) for f in os.listdir(tfrecord_dir) if f.endswith(".tfrecord")]
def parse_function(example_proto):
    features = {
    'image/encoded': tf.io.FixedLenFeature([], tf.string, default_value=''),
    'image/width': tf.io.FixedLenFeature([], tf.int64, default_value=0),
    'image/height': tf.io.FixedLenFeature([], tf.int64, default_value=0),
    'top_category_id': tf.io.FixedLenFeature([], tf.int64, default_value=0),
    'color_text': tf.io.FixedLenFeature([], tf.string, default_value=''),
    'top_category_text': tf.io.FixedLenFeature([], tf.string, default_value=''),
    'bottom_category_text': tf.io.FixedLenFeature([], tf.string, default_value='')
    }
    parsed_features = tf.io.parse_single_example(example_proto, features)
    image = tf.image.decode_jpeg(parsed_features['image/encoded'], channels=3)
    image =tf.cast(image, tf.float32) / 255.0
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, [112, 112]) 
    image = tf.reshape(image, [112, 112, 3])
    label = tf.one_hot(parsed_features['top_category_id'], depth=15)
    
    return image, label


# Get file paths
# tfrecord_filename_train = "data/tfrecords/train/{tfrecord_filenames_train[0]}"
#Storing raw data as TFRecordDataset
raw_dataset_train = tf.data.TFRecordDataset(tfrecord_filenames_train)
#Creating a dataset with the parsed features using map
dataset = raw_dataset_train.map(parse_function, num_parallel_calls=tf.data.experimental.AUTOTUNE)
```

```python
def get_dataset_partitions_tf(dataset, ds_size, label_counts, train_split=0.8, val_split=0.2, shuffle=True, shuffle_size=500):
    assert (train_split + val_split) == 1
    
    if shuffle:
        # Specify seed to always have the same split distribution between runs
        dataset = dataset.shuffle(shuffle_size, seed=12)
    
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    # Calculate the number of samples to take from each class for training and validation
    train_label_counts = [int(count * train_split) for count in label_counts]
    val_label_counts = [count - train_count for count, train_count in zip(label_counts, train_label_counts)]
    # Select the first 50,000 images
    dataset = dataset.take(10000)
    # Filter the dataset by class label
    train_datasets = []
    val_datasets = []
    for i in range(len(label_counts)):
        class_ds = dataset.filter(lambda x, y: tf.equal(tf.argmax(y), i))
        train_ds_class = class_ds.take(train_label_counts[i]).shuffle(buffer_size=shuffle_size).repeat()
        val_ds_class = class_ds.skip(train_label_counts[i]).take(val_label_counts[i])
        train_datasets.append(train_ds_class)
        val_datasets.append(val_ds_class)
    
    # Combine the filtered datasets into a single dataset for training and validation
    train_ds2 = train_datasets[0]
    val_ds2 = val_datasets[0]
    for i in range(1, len(train_datasets)):
        train_ds2 = train_ds2.concatenate(train_datasets[i])
        val_ds2 = val_ds2.concatenate(val_datasets[i])
    
    return train_ds2, val_ds2
```

```python
train_ds2, val_ds2 = get_dataset_partitions_tf(dataset, 10000, label_counts, train_split=0.8, val_split=0.2, shuffle=True, shuffle_size=5000)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import GaussianNoise
batch_size=32
# Define the image augmentation pipeline

def augment(image, label):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.75, upper=1.25)
    return image, label

# Apply data augmentation to the dataset
train_ds_augmented = train_ds2.map(augment)
```

```python
from tensorflow.keras.applications import EfficientNetB0
lr_decay = 0.1
lr_patience = 3
learningrate = 0.001

policy = tf.keras.mixed_precision.Policy('mixed_float16')
tf.keras.mixed_precision.set_global_policy(policy)

pretrained_model =EfficientNetB0(weights='imagenet', include_top=False, input_shape=(112, 112, 3))

# Freeze layers in base_model.layers to prevent them from being trained
for layer in pretrained_model.layers:
    layer.trainable = False

# Add more layers to the model
model = Sequential([
    pretrained_model,
    GlobalAveragePooling2D(),
    Dropout(0.2),
    Dense(15, activation='softmax', dtype='float32', kernel_regularizer=l2(0.001))
])

# Define early stopping callback
early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=lr_decay, patience=lr_patience, verbose=1, min_lr=1e-7)

# Compile model with mixed precision optimizer
optum = Adam(learning_rate=learningrate)
model.compile(optimizer=optum, loss="categorical_crossentropy", metrics=["accuracy"])

with tf.device('/device:GPU:0'):
    hist = model.fit(
        train_ds_augmentedbatch,
        epochs=20,
        steps_per_epoch = steps,
        validation_data=val_ds_batch,
        validation_steps = val_steps,
        callbacks=[early_stop, reduce_lr,checkpoint,tensorboard_callback],
        class_weight=class_weights,
        verbose=1
    )
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Make predictions on the validation set
y_pred = model.predict(val_ds_batch)

# Convert predictions to class labels
y_pred = np.argmax(y_pred, axis=1)

# Get actual class labels
y_true = val_ds_batch.classes

# Calculate confusion matrix
cm = confusion_matrix(y_true, y_pred)

print(cm)
```

```python
from sklearn.metrics import f1_score


# Calculate F1-score for each class
f1 = f1_score(y_true, y_pred, average=None)

# Calculate macro-averaged F1-score (weighted average of F1-scores for each class)
macro_f1 = f1_score(y_true, y_pred, average='macro')

# Calculate micro-averaged F1-score (based on total number of true positives, false positives and false negatives)
micro_f1 = f1_score(y_true, y_pred, average='micro')

# Print F1-score for each class, as well as macro- and micro-averaged F1-score
print("F1-score for each class:", f1)
print("Macro-averaged F1-score:", macro_f1)
print("Micro-averaged F1-score:", micro_f1)
```
